### Mackey-Glass Time Series

```python
import numpy as np
import matplotlib.pyplot as plt
from rc_datasets import rc_datasets
```

The time-series is defined as: 
$$\frac{dx(t)}{dt}=\frac{ax(t-\tau)}{1+x(t-\tau)^{10}}-bx(t)$$
where $a, b,\tau$ are parameter. The typical parameters used to generate a pseudo-chaotic time-series are:
$$ a=0.2, b=0.1, \tau=17 $$

#### Numerical implementation
Numerically, we solve the evolution equations for $x(t)$ using a Runge-Kutta algorithm.

$x_0$ is our initial condition, $dt$ is the time-step of the evolution, and $N_t$ is the total number of points. 

I have included an interpolation factor $fac$ that is normally used in the literature. 

```python
fac=10
Ttot=1000
Nt=fac*Ttot
x0=1.8
dt=0.1

ds=rc_datasets.mackeyglass(fac,Nt,dt,x0,0.2,0.1,17)
```

With this definition, $ds$ is a Python object containing two variables: ds.u and ds.t.
